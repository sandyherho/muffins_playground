```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
colorbar_levels = np.arange(-7, 7)
```

```python
plt.fill_between(amoc_2e17_anom["lat_moc"], 
                 amoc_2e17_anom["zt_moc"].min(), 
                 amoc_2e17_anom["zt_moc"].max(), 
                 color='#695447')

contour_filled = plt.contourf(amoc_2e1_anom["lat_moc"], 
                              amoc_2e1_anom["zt_moc"],
                              amoc_2e1_anom, 
                              levels=colorbar_levels,
                              cmap="RdBu_r")

contour_lines = plt.contour(amoc_2e1_anom["lat_moc"], 
                            amoc_2e1_anom["zt_moc"], 
                            amoc_2e1_anom, 
                            levels=colorbar_levels,
                            colors='k', linewidths=0.5)

plt.xlabel("Latitude [°]")
plt.ylabel("Depth [m]")

cbar = plt.colorbar(contour_filled)
cbar.set_label("Anomaly [Sv]")
plt.clabel(contour_lines, inline=True, fontsize=10, fmt='%1.1f')

plt.gca().invert_yaxis()

plt.title(r"AMOC $\Psi$ anomaly under $2\times 10^{1}$ Sv freshwater hosing")

plt.savefig("../figs/amoc_2e1.png", dpi=300)
```
